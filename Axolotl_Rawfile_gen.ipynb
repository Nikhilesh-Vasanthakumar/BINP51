```python
# Set the barcode file
#Summarizing the loop
file_path = "/home/nikvaku/snic2022-6-312/LabMemberScratchDir/Nikhilesh/Inter_data/Axolotl_barcodes.txt"

# Open the file and read the single line
with open(file_path, 'r') as file:
    # Read the single line from the file
    line = file.readline()

# Split the line by tabs, remove the last two characters from each element, and create a list
barcode_list = [element[:-2] for element in line.strip().split('\t')]
# Set the directory containing filtered matrices
filtered_dir = "/home/nikvaku/snic2022-6-312/LabMemberScratchDir/Nikhilesh/Raw_data/GSE201446_RAW/Extracted_NEO"

# Initialize an AnnData object
combined_data = None
for filename in os.listdir(filtered_dir):
    mat = sc.read(os.path.join(filtered_dir, filename))
    mat_T=mat.transpose()
    cells=mat.var
    mat_T.var=mat.obs
    mat_T.obs=cells
    mat_T = mat_T[mat_T.obs.index.isin(barcode_list)]
    if combined_data is None:
        combined_data = mat_T
    else:
        combined_data = anndata.concat([combined_data,mat_T], join='outer')
```
